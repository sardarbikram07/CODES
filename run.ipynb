```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.widgets import Button

fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

circle = plt.Circle((0, 5), 0.5)
ax.add_patch(circle)

paused = False

def update(frame):
    if not paused:
        circle.center = (frame % 10, 5)
    return circle,

ani = animation.FuncAnimation(
    fig, update,
    frames=100,
    interval=200
)

# ----- Buttons -----
ax_pause = plt.axes([0.3, 0.05, 0.15, 0.075])
ax_play = plt.axes([0.55, 0.05, 0.15, 0.075])

btn_pause = Button(ax_pause, "Pause")
btn_play = Button(ax_play, "Play")

def pause(event):
    global paused
    paused = True

def play(event):
    global paused
    paused = False

btn_pause.on_clicked(pause)
btn_play.on_clicked(play)

plt.show()
```

```python
import matplotlib.pyplot as plt

n = int(input("Enter number of slices: "))

labels = []
sizes = []
for i in range(n):
    label = input(f"Enter label {i+1}: ")
    size = float(input(f"Enter size for {label}: "))
    labels.append(label)
    sizes.append(size)
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart (User Input)")

plt.show()
```

```python
import matplotlib.pyplot as plt

n = int(input("Enter number of categories: "))

categories = []
values = []

for i in range(n):
    cat = input(f"Enter category {i+1}: ")
    val = int(input(f"Enter value for {cat}: "))
    categories.append(cat)
    values.append(val)

plt.bar(categories, values)
plt.title("Bar Chart (User Input)")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for line chart
x1 = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 25, 30]

# Data for scatter plot
x2 = [1, 2, 3, 4, 5]
y2 = [8, 18, 12, 28, 22]

plt.figure(figsize=(10, 5))

# 🔹 Line Chart
plt.subplot(1, 2, 1)
plt.plot(x1, y1, marker='o')
plt.title("Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# 🔹 Scatter Plot
plt.subplot(1, 2, 2)
plt.scatter(x2, y2)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.tight_layout()
plt.show()
```
